```python
import requests
import re
import pandas as pd
import os
```

# Settings

```python
start_date = "2022-09-01"
end_date = "2022-10-25"
sensor_ids = ["74650","67867"]
url_sc = 'https://archive.sensor.community'

# https://sensor.community/en/
```

# Get data

## Get the file names for all the sensors for the first date (this is slow...)

```python
dates = pd.date_range(start_date,end_date,freq='d')

#date = dates[0]
#sensor_id = sensor_ids[0]

sensor_id_files = []

date = str(dates[0].date())
url = "/".join([url_sc,date])
r1 = requests.get(url)
data = r1.text
for sensor_id in sensor_ids:   
    try:
        start = re.escape(date)
        #start = re.escape(f'{sensor_id}.csv">')
        end   = re.escape(f'{sensor_id}.csv') 
        result = re.search('%s(.*)%s' % (start, end), data).group(0)
        result=result.split(">")[1]
        print(result)
        sensor_id_files.append(result)

        #url_csv = "/".join([url, result])        
        #df2 = pd.read_csv(url_csv, sep=";")
        #if type(df) == int:
        #    df = df2
        #else:
        #    df = pd.concat([df, df2])
    except:
        print(f"Something went wrong for {date} and sensor {sensor_id}")
```

```python
df = -1

for date in dates:
    date = str(date.date())
    
    url = "/".join([url_sc,date])
  
    for sensor_csv in sensor_id_files:  
        old_date = sensor_csv.split("_")[0]
        sensor_csv = sensor_csv.replace(old_date, date)
        
        try:    
            url_csv = "/".join([url, sensor_csv])
            
            df2 = pd.read_csv(url_csv, sep=";")
            if type(df) == int:
                df = df2
            else:
                df = pd.concat([df, df2], ignore_index=True)
        except:
            print(f"Something went wrong for {date} and sensor {sensor_csv}")
        
        
```

```python
df.to_csv(os.path.join("data", "sc_export.csv"), index=False)
```
