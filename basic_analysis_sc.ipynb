# Read the data

```python
df = pd.read_csv(os.path.join("data", "sc_weather.csv"))
```

```python
df.columns
```

# Weekend

```python
sns.catplot(data=df, y='P1', x='sensor_id', hue="weekend", kind="violin", palette = flatui, cut=0)
f = plt.gcf()
f.set_size_inches(15,10)
plt.show()
```

```python
sns.catplot(data=df, y='P2', x='sensor_id', hue="weekend", kind="violin", palette = flatui, cut=0)
f = plt.gcf()
f.set_size_inches(15,10)
plt.show()
```

# Hourly distribution

```python
for sensor_id in df.sensor_id.unique():
    sns.scatterplot(data=df[df.sensor_id == sensor_id], y='P1', x='time', s=15, alpha=0.5, palette = flatui[:2])

    f = plt.gcf()
    f.set_size_inches(15,10)
    plt.xlabel("Time of day")
    plt.ylabel("P1")
    plt.title(f"P1({sensor_id})")
    
    
    plt.xticks(np.arange(24), np.arange(24))
    #plt.legend(routes.astype(str))
    plt.show()
```

```python
for sensor_id in df.sensor_id.unique():
    sns.scatterplot(data=df[df.sensor_id == sensor_id], y='P2', x='time', s=15, alpha=0.5, palette = flatui[:2])

    f = plt.gcf()
    f.set_size_inches(15,10)
    plt.xlabel("Time of day")
    plt.ylabel("P2")
    plt.title(f"P2({sensor_id})")
    
    
    plt.xticks(np.arange(24), np.arange(24))
    #plt.legend(routes.astype(str))
    plt.show()
```
